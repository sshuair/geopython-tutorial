重新投影

```python
from PIL import Image
```

```python
import numpy as np
import rasterio
from rasterio.warp import calculate_default_transform, reproject, Resampling
from rasterio import crs

rgb = 'images/world.rgb.tif'
out = 'images/reproj.tif'

dst_crs = crs.CRS = "EPSG:3857"




with rasterio.open(rgb) as src:
    #获取原始数据的配置 投影坐标系 长宽 
    profile = src.profile
    print(profile)
    # 计算转换后的仿射变换和长宽   按照原始数据的配置和 目标投影坐标系
    dst_affine, dst_width, dst_height = calculate_default_transform(
        src.crs, dst_crs, src.width, src.height, *src.bounds)

    # 更新配置
    profile.update({
        'crs': dst_crs,
        'transform': dst_affine,
        #'affine': dst_affine,
        'width': dst_width,
        'height': dst_height
    })

    

    with rasterio.open(out, 'w', **profile) as dst: #写入到新的文件中
        for i in range(1, src.count +1):
            src_array = src.read(i) #读取每个波段数据
            dst_array = np.empty((dst_height, dst_width), dtype='uint8')
            
            #重投影
            reproject(
                
                source=src_array,
                src_crs=src.crs,
                src_transform=src.affine,
                # Destination paramaters
                destination=dst_array,
                dst_transform=dst_affine,
                dst_crs=dst_crs,
                # Configuration
                resampling=Resampling.nearest,
                num_threads=2)
            
            dst.write(dst_array, i)
```
